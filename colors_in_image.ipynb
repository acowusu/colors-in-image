# Full working script
note this also requres the color definitions art the top of this notebook

```
from matplotlib import pyplot as plt
import cv2

bgrimage = cv2.imread('/content/test.png')
rgbimage = cv2.cvtColor(bgrimage, cv2.COLOR_RGB2BGR)

flattened = rgbimage.reshape(rgbimage.shape[0]*rgbimage.shape[1],rgbimage.shape[2])

avgcolor = np.median(flattened, axis = 0) # this is the average color for the whole image

print(closestcolor(avgcolor))
```

# Wait we may have an issue!
our current approach has a major flaw : an image which is half black and half white whould be considered grey.
 

 lets plot a scatter plot or the pixels in our image to  better visulase the colors we have

```
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(flattened[:,0], flattened[:,1], flattened[:,2], c='r', marker='o')

ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')

plt.show()

```

It looks like there are multiple colors within this image why don't we get the main ones 

# Getting the Donminent N colors

my first thought:

1.   calculate the color name for each pixel
2.   findthe n most common names

I disregarded this as  pixels similar in value could have different names if they are on the edge of a cluster. in addition you may have slightly different pixles that look like a single block of color whn you zoom out.
Note: i amnot just and then  .

As a result I am going to use an algorithm called k  means clustering to group the colors and then get the color name for the center of the cluster


```
from sklearn.cluster import KMeans
clusters = 3 #@param {type: "number"} 
model = KMeans(n_clusters=clusters)
model.fit(flattened)
model.cluster_centers_

for color in model.cluster_centers_:
  print(closestcolor(color))
```
